```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.linear_model import LinearRegression,Lasso
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('/workspaces/Daily-machineLearning/Gold_Price_Prediction/Dataset/gld_price_data.csv')
df.head()
```

```python
print(df.shape[0],df['Date'].nunique())
```

```python
df.drop(columns=['Date'],inplace=True)
```

```python
df.corr()
```

```python
sc=StandardScaler()
```

```python
features=df.drop(columns=['GLD'],axis=1)
target=df['GLD']
features=sc.fit_transform(features)
# target=sc.fit_transform([target])
```

```python
# target=target.flatten()
```

```python
target
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,target,test_size=0.2,random_state=2)
x_train
```

```python
model1=LinearRegression()
model1.fit(x_train,y_train)
```

```python
y_pred=model1.predict(x_test)
print('mean square error:',mean_squared_error(y_pred,y_test))
print('r2 score:',r2_score(y_pred,y_test))
```

```python
model2=Lasso(alpha=0.1)
model2.fit(x_train,y_train)
```

```python
y_pred=model2.predict(x_test)
print('mean square error:',mean_squared_error(y_pred,y_test))
print('r2 score:',r2_score(y_pred,y_test))
```

```python
print(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model=RandomForestRegressor(n_estimators=100,max_depth=5,random_state=2)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
print('mean square error:',mean_squared_error(y_pred,y_test))
print('r2 score:',r2_score(y_pred,y_test))
```

```python
print(y_test,pd.Series(y_pred))
```
