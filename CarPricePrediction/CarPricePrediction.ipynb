```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.linear_model import LinearRegression,Lasso
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('/workspaces/Daily-machineLearning/CarPricePrediction/Dataset/car data.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
print(df['Car_Name'].value_counts())
print(df['Fuel_Type'].value_counts())
print(df['Seller_Type'].value_counts())
print(df['Transmission'].value_counts())
```

<h1>Encoding and Normalizing data</h1>

```python
le=LabelEncoder()
sd=StandardScaler()
cat=['Fuel_Type','Seller_Type','Transmission']
num=['Year','Selling_Price','Kms_Driven','Owner']
for i in cat:
    df[i]=le.fit_transform(df[i])
df[num]=sd.fit_transform(df[num])
```

<h2>Features and Target separation</h2>

```python
features=df.drop(columns=['Car_Name','Present_Price'],axis=1)
target=df['Present_Price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,target,test_size=0.2,random_state=2)
x_train
```

<h1>Traing and Testing Models</h1>

```python
model1=LinearRegression()
model1.fit(x_train,y_train)
```

```python
y_pred=model1.predict(x_test)
print('mean square error:',mean_squared_error(y_pred,y_test))
print('r2 score:',r2_score(y_pred,y_test))
```

```python
model2=Lasso(alpha=0.1)
model2.fit(x_train,y_train)
```

```python
y_pred=model2.predict(x_test)
print('mean square error:',mean_squared_error(y_pred,y_test))
print('r2 score:',r2_score(y_pred,y_test))
```
